```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df1=MinMaxScaler().fit_transform(df[['album','track_number','acousticness', 'danceability', 'energy', 'instrumentalness',
      'liveness', 'loudness', 'speechiness', 'tempo', 'valence', 'popularity',
     'duration_ms']])
```

```python
df1=pd.DataFrame(df1,columns=['album','track_number','acousticness', 'danceability', 'energy', 'instrumentalness',
      'liveness', 'loudness', 'speechiness', 'tempo', 'valence', 'popularity',
     'duration_ms'])
```

### Model building 

```python
X=df1.drop(columns='popularity')
Y=pd.DataFrame(df1['popularity'])
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=123)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from boruta import BorutaPy
```

```python
RFC1=RandomForestRegressor(n_estimators=1000,max_depth=5,random_state=1)
boruta_selector=BorutaPy(RFC1, n_estimators='auto',random_state=1,verbose=2)
```

```python
boruta_selector.fit(np.array(X_train),np.array(Y_train))
```

```python
selected_features=pd.DataFrame({'Features:':list(X_train.columns),'Ranking':boruta_selector.ranking_})
```

```python
selected_features
```

```python
X_train_boruta=boruta_selector.transform(np.array(X_train))
X_test_boruta=boruta_selector.transform(np.array(X_test))
```

```python
RFR=RandomForestRegressor(n_estimators=1000,max_depth=5,random_state=1)
RFR.fit(X_train_boruta,Y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,RFR.predict(X_test_boruta))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
LR.fit(X_train,Y_train)
```

```python
r2_score(Y_test,LR.predict(X_test))
```

```python
#Energy,danceability,tempo have high VIF let's remove it
X_new=X.drop(columns=['tempo','energy','danceability'])
```

```python
X_train1,X_test1,Y_train1,Y_test1=train_test_split(X_new,Y,test_size=0.8,random_state=123)
```

```python
#fit linear regression after removing columns
LR1=LinearRegression()
LR1.fit(X_train1,Y_train1)
```

```python
r2_score(Y_test1,LR1.predict(X_test1))
#r2 score is less than previous one after removing columns which has high VIF
```

```python
from sklearn.cluster import KMeans
```

```python
cs = []  # Initialize an empty list to store inertia values

for i in range(2, 10):  # Loop over the range 1 to 9 (inclusive) to evaluate different numbers of clusters
    K = KMeans(n_clusters=i, init='k-means++', max_iter=100)  # Create a KMeans instance with i clusters
    K.fit(X)  # Fit the KMeans model to the data X
    cs.append(K.inertia_)  # Append the inertia value (sum of squared distances of samples to their closest cluster center) to the list
```

```python
plt.plot(range(2,10),cs)
plt.xlabel('n_clusters')
plt.ylabel('cs')
plt.title('Elbow method')
```

```python
K=KMeans(3, init='k-means++', max_iter=100,random_state=3)
```

```python
K.fit(X_train,Y_train)
```

```python
df['KM_Labels'] = K.fit_predict(X)
```

```python
df['KM_Labels'].value_counts()
```

```python
Y=pd.Series(df['popularity'])
```

```python
sum(Y==K.labels_)
```
